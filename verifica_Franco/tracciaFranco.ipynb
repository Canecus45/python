# **Gestione dei referti medici**

## Classe RefertoMedico:

Crea una classe RefertoMedico in grado di contenere il referto medico di un paziente di cui si fornisce il nome.
Ogni referto medico sarà composto da una lista di tuple, ogni tupla è una diagnosi contenente (data rilevazione, valore rilevato, valore di riferimento, nome del parametro rilevato). In date diverse possono essere presenti più rilevazioni dello stesso parametro.
esempio:


````python
[("01/03/2022",323,350,"parametro1"),("01/03/2022",183,150,"parametro2"),("01/05/2023",163,150,"parametro2"),("01/06/2024",137,150,"parametro2"),...,]
````

La classe RefertoMedico deve avere i seguenti metodi:


1. popola(): per inserire le tuple con le varie rilevazioni, utilizza dati statici senza fare input manuale.
2. visualizzaReferto(): per visualizzare il referto medico inserito, stampando singolarmente e su righe diverse i valori delle tuple: esempio:
 ```
 Data:01/03/2022
 Parametro: parametro1
 Valore rilevato:323
 Valore di riferimento:350
 Scostamento: |350-323|=27

 Data:01/03/2022
 Parametro: parametro2
 Valore rilevato:183
 Valore di riferimento:150
 Scostamento: |150-183|=23
 ....
```
3. contaDiagnosi() : Conta e visualizza il numero di diagnosi presenti nel referto medico, per visualizzare le diagnosi richiama il metodo precedente.
4. mediaRilevazioni(parametro): Dato un parametro, calcola e visualizza la media delle rilevazioni (per i parametri che compaiono più volte).
5. rilevazioneMigliore(parametro) : Individua e visualizza le migliori rilevazioni presenti, cioè quelle che per un dato parametro hanno il minimo valore di scostamento(differenza in abs) tra il valore di riferimento e il valore rilevato. Restituisce un dizionario avente come chiave il nome del parametro e come valori una tupla contenente il valore di riferimento, lo scostamento tra il valore rilevato e il valore di riferimento, la lista delle date delle rilevazioni.
Esempio del dizionario restituito :


```{"parametro2": (150,13,["01/05/2023","01/06/2024"])}```

Esempio dell'output da produrre :
```
Migliori rilevazioni di: parametro2
Valore di riferimento: 150
Scostamento minimo registrato : 13
Nelle rilevazioni del : 01/05/2023,01/06/2024
```
